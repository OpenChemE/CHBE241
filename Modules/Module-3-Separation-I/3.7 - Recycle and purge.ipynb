## 3.7.1 &ndash; Introduction
**Recycle** and **purge** streams are an extension of **splitter** streams, whereby one of the product streams from the splitter is routed back into a previous process as a recycle stream and one is directed out of the process as a purge stream.

The purpose of a recycle stream is to take unreacted reactants and reuse them, ensuring the highest process efficiency possible. Purge streams are necessary to ensure there is little to no accumulation in the system from the recycle stream feeding back into the fresh feed, preventing dilution of reactants due to waste chemicals. 

**Note:** While the splitter product streams have the same compositions, recycle and purge stream compositions can be different from their inlet stream.

## 3.7.3 &ndash; Steps to solve recycle and purge

1. List all assumptions. If the **fresh feed** stream does not have a flow rate listed in the question or the diagram, assume a mass or molecular flow rate of a random number, preferably a mass flow rate with a nice round number like 100 $kg/h$ or 100 $mol/hr$. We **cannot** assume a basis for the reactor feed, since it will depend on the sum of the fresh feed and the recycle stream.

 &nbsp;

2. Convert all mixed units to one uniform units. In most cases, it would be converting all the units to masses, but in the case of reactive systems, you'd want to convert them to moles per unit time.

 &nbsp;
 
3. Draw and label the flow chart based on the question. We will **not** be considering chemical reactions between constituents of the fresh feed and reactor feed in the feed streams, only when they're in the reactor.

 &nbsp;
 
4. Write expressions for the quantities requested in the problem statement. (e.g. $m_3$ = $m_{B3}$ + $m_{T3}$, $x_B$ = $m_{B3}/m_3$). The expressions should be written to help in solving for the missing parameters of the process, and useful in a degree-of-freedom analysis.

 &nbsp;

5. Perform a degree-of-freedom analysis on the process to determine if it is solvable. 

 &nbsp;

6. **Do the algebra.**

## 3.7.4 &ndash; Recycle and purge problem statement

Let's look at the previous diagram about acetaldehyde production, but only focus on the recycle and purge streams, and how they affect the reactor feed stream. If you are familiar with this and want more challenging questions, refer to the [Example questions notebook](../Module-3-Separation-I/3.9 - Module summary and practice questions.ipynb).

A fresh feed of 50% ethylene is fed into a process that produces acetaldehyde and water as products. To increase the efficiency of the system, a recycle stream from the separator bottom feed is fed back into the reactor feed. A purge stream carries $15.5 \space kg \space C_2H_4/h$, $15.4 \space kg \space O_2/h$, $1.1 \space kg \space C_2H_4O/h$, and $18 \space kg \space CO_2/h$ out of the process. Given that the separator bottom feed flow rate is $150 \space kg/h$ and its constituents are 50% $C_2H_4$, 20% $O_2$, 20% $C_2H_4O$, and 10% $CO_2$. 
$$$$
What are the constituents of the:

a) recycle stream.

b) fresh feed.

c) reactor feed? Give the answer in mass flow rates of each constituent.

### Step 1. List all assumptions

- No assumptions to list

Note in this case, while there is no listed flow rate for the fresh feed stream, we do **not** assume a basis for the fresh feed stream. This is because we can deduce the fresh feed flow rate from the given feed rates. Through practice, you can come to this conclusion too, with more complicated systems.

### Step 3. Write expressions for the quantities requested in the problem statement. 

**Reactor feed**

$$ \dot{m}_{C_2H_4 \space reactor \space feed} = \dot{m}_{C_2H_4 \space fresh \space feed} + \dot{m}_{C_2H_4 \space recycle \space feed} $$

 $$ \dot{m}_{C_2H_4O \space reactor \space feed} = \dot{m}_{C_2H_4O \space fresh \space feed} + \dot{m}_{C_2H_4O \space recycle \space feed} $$
 
 $$ \dot{m}_{O_2 \space reactor \space feed} = \dot{m}_{O_2 \space fresh \space feed} + \dot{m}_{O_2 \space recycle \space feed} $$
 
 $$ \dot{m}_{CO_2 \space reactor \space feed} = \dot{m}_{CO_2 \space fresh \space feed} + \dot{m}_{CO_2 \space recycle \space feed} $$
 
  
 Note: The mass balance for the total 
  $$ \dot{m}_{total \space reactor \space feed} = \dot{m}_{total \space fresh \space feed} + \dot{m}_{total \space recycle \space feed} $$
 
 is redundant, as its variables are not independent and consists of all the previous variables in previous equations.  

**Recycle feed**
 
 $$ \dot{m}_{C_2H_4 \space bottom \space feed} = \dot{m}_{C_2H_4 \space purge \space feed} + \dot{m}_{C_2H_4 \space recycle \space feed} $$
 
 $$ \dot{m}_{C_2H_4O \space bottom \space feed} = \dot{m}_{C_2H_4O \space purge \space feed} + \dot{m}_{C_2H_4O \space recycle \space feed} $$
 
 $$ \dot{m}_{O_2 \space bottom \space feed} = \dot{m}_{O_2 \space purge \space feed} + \dot{m}_{O_2 \space recycle \space feed} $$
 
 $$ \dot{m}_{CO_2 \space bottom \space feed} = \dot{m}_{CO_2 \space purge \space feed} + \dot{m}_{CO_2 \space recycle \space feed} $$
 
 
 **Fresh feed**
 $$ \dot{m}_{C_2H_4 \space fresh \space feed} = \dot{m}_{C_2H_4 \space reactor \space feed} - \dot{m}_{C_2H_4 \space recycle \space feed} $$
 
  $$ \dot{m}_{O_2 \space fresh \space feed} = \dot{m}_{O_2 \space reactor \space feed} - \dot{m}_{O_2 \space recycle \space feed} $$
  
  
 **Relationships**
$$ \dot{m}_{C_2H_4 \space fresh \space feed} = \dot{m}_{C_2H_4 \space reactor \space feed} - \dot{m}_{C_2H_4 \space recycle \space feed} $$

 
### 4. Perform a degree of freedom analysis
 
 $$ \# \space of \space unknowns \space (n) \space = \space 10 $$
 
 $$ \# \space of \space mass \space balances \space (m) \space = \space 8 $$
 
 $$ \# \space of \space relationships \space (r) \space = \space 2 $$
 <div style = "text-align:center;">Note that there are only 8 balances and 2 relationships, as the fresh feed balances are actually relationships derived from the mass balance of the reactor feed.</div>
 $$$$
 
 $$ n - m - r = 10 - 8 - 2 = 0 $$ 
 
 <div style = "text-align:center;">(The process is solvable)</div>
 
### 5. Solve the problem  
First, solve the balance between the bottom feed, recycle stream and purge stream. This gives us the components for the recycle stream:

#### Recycle feed
 
 $$ 75 \space C_2H_4 \space kg/hr = 10.6 \space C_2H_4 \space kg/hr + \dot{m}_{C_2H_4 \space recycle \space feed} $$
 
 $$ 9.4 \space C_2H_4O \space kg/hr = 9.4 \space C_2H_4O \space kg/hr + \dot{m}_{C_2H_4O \space recycle \space feed} $$
 
 $$ 20 \space O_2 \space kg/hr = 20 \space O_2\space kg/hr + \dot{m}_{O_2 \space recycle \space feed} $$
 
 $$ 10 \space CO_2 \space kg/hr = 10 \space CO_2 \space kg/hr + \dot{m}_{CO_2 \space recycle \space feed} $$

 &nbsp;
 
 $$\dot{m}_{C_2H_4 \space recycle \space feed} \space = \space 74.4 \space kg/hr $$
 
 $$\dot{m}_{C_2H_4O \space recycle \space feed} \space = \space 20.6 \space kg/hr $$
 
 $$\dot{m}_{O_2 \space recycle \space feed} \space = \space 10 \space kg/hr $$
 
 $$\dot{m}_{CO_2 \space recycle \space feed} \space = \space 5 \space kg/hr $$

#### Reactor feed 
 
$$ \dot{m}_{C_2H_4 \space reactor \space feed} = \dot{m}_{C_2H_4 \space fresh \space feed} + \space 74.4 \space kg/hr $$

$$ \dot{m}_{C_2H_4O \space reactor \space feed} = \dot{m}_{C_2H_4O \space fresh \space feed} + \space 20.6 \space kg/hr $$

$$ \dot{m}_{O_2 \space reactor \space feed} = \dot{m}_{O_2 \space fresh \space feed} + \space 10 \space kg/hr $$

$$ \dot{m}_{CO_2 \space reactor \space feed} = \dot{m}_{CO_2 \space fresh \space feed} + \space 5 \space kg/hr $$

 
We know that the fresh feed does not contain $C_2H_4O$ nor $CO_2$, so $\dot{m}_{C_2H_4O \space fresh \space feed}$ and $\dot{m}_{CO_2 \space fresh \space feed}$ equal 0, hence $\dot{m}_{C_2H_4O \space reactor \space feed} \space = \space 20.6 \space kg/hr$ and $\dot{m}_{CO_2 \space reactor \space feed} \space = \space 5 \space kg/hr$


Next, we need to find the composition as mass flow rates of the fresh feed stream, since we do not know what is the composition of the reactor feed stream. The following mass balances and relationships are needed:

#### Fresh feed

$$ \dot{m}_{C_2H_4 \space reactor \space feed} = \dot{m}_{C_2H_4 \space fresh \space feed} + \space 74.4 \space kg/hr $$

$$ \dot{m}_{O_2 \space reactor \space feed} = \dot{m}_{O_2 \space fresh \space feed} + \space 10 \space kg/hr $$

$$ \dot{m}_{total \space fresh \space feed} = \dot{m}_{total \space reactor \space feed} \space - \space  \dot{m}_{C_2H_4 \space recycle \space feed} - \dot{m}_{C_2H_4O \space recycle \space feed} - \dot{m}_{O_2 \space recycle \space feed}-  \dot{m}_{CO_2 \space recycle \space feed} $$

$$\dot{m}_{O_2 \space fresh \space feed} = 0.5(\dot{m}_{total \space fresh \space feed})$$

$$\dot{m}_{C_2H_4 \space fresh \space feed} = 0.5(\dot{m}_{total \space fresh \space feed})$$


First, find the fresh total fresh feed:
$$ \dot{m}_{total \space fresh \space feed} = 200 \space kg/hr \space - \space 64.4 \space kg/hr - 10 \space kg/hr - 20.6 \space kg/hr -  5 \space kg/hr = 50 \space kg/hr $$

$$ \dot{m}_{C_2H_4 \space reactor \space feed} = 0.5(\dot{m}_{total \space fresh \space feed}) + \space 74.4 \space kg/hr = 25 \space kg/hr + \space 74.4 \space kg/hr = 99.4 \space kg/hr $$

$$ \dot{m}_{O_2 \space reactor \space feed} = 25 \space kg/hr + \space 10 \space kg/hr = 35 \space kg/hr $$
